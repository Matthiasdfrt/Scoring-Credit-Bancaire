## 1. Import des librairies et chargement des données

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

pd.set_option('display.max_columns', 100)
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
path = 'DATA'
filename = 'application_train_cleaned.csv'

print(f"Chargement de {filename}...")
df = pd.read_csv(os.path.join(path, filename))
print(f"Dimensions du dataset : {df.shape}")
df.head()
```

## 2. Analyse Univariée : La Variable Cible (TARGET)

```python
target_counts = df['TARGET'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 6))
sns.barplot(x=target_counts.index, y=target_counts.values, palette=['#3B82F6', '#EF4444'])
plt.title('Distribution de la Variable Cible (0: Rembourse, 1: Défaut)')
plt.xlabel('Target')
plt.ylabel('Pourcentage (%)')
plt.text(0, target_counts[0], f"{target_counts[0]:.1f}%", ha='center', va='bottom')
plt.text(1, target_counts[1], f"{target_counts[1]:.1f}%", ha='center', va='bottom')
plt.show()

print(f"Déséquilibre de classe : {target_counts[1]:.2f}% de défauts.")
```

## 3. Analyse des Corrélations avec la Target
Quelles sont les variables les plus corrélées (positivement ou négativement) avec le défaut de paiement ?

```python
correlations = df.corr(numeric_only=True)['TARGET'].sort_values()

print("Top 10 Corrélations Positives (Facteurs de risque) :")
print(correlations.tail(11).drop('TARGET').sort_values(ascending=False))

print("\nTop 10 Corrélations Négatives (Facteurs de fiabilité) :")
print(correlations.head(10))
```

## 4. Analyse Bivariée : Variables Clés
Visualisation de la densité des variables les plus importantes selon la classe (0 ou 1).

```python
def plot_kde_target(var_name, df):
    plt.figure(figsize=(10, 6))
    sns.kdeplot(df.loc[df['TARGET'] == 0, var_name], label='Target 0 (Rembourse)', color='#3B82F6', fill=True, alpha=0.1)
    sns.kdeplot(df.loc[df['TARGET'] == 1, var_name], label='Target 1 (Défaut)', color='#EF4444', fill=True, alpha=0.1)
    plt.title(f'Distribution de {var_name} selon la Target')
    plt.xlabel(var_name)
    plt.ylabel('Densité')
    plt.legend()
    plt.show()

plot_kde_target('EXT_SOURCE_2', df)
plot_kde_target('EXT_SOURCE_3', df)
```

```python
df['YEARS_BIRTH'] = abs(df['DAYS_BIRTH']) / 365
plot_kde_target('YEARS_BIRTH', df)
```

```python
plot_kde_target('PAYMENT_RATE', df)
```
